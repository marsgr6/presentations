```python
import matplotlib.pyplot as plt
import numpy as np

# Create random walk
d = 10  # directions
theta = np.linspace(0, 2*np.pi, d, endpoint=False)
r = 1  # radius

steps = 100

walks = 100000

def make_walk(steps, r=None, plot=False):
    if r is None: r = 1
    else: r = np.random.randint(1, 6, size=steps) + np.random.randint(1, 11, size=steps)
    z = (r * np.exp(1j * np.random.choice(theta, steps))).cumsum()

    x = z.real
    y = z.imag

    if plot: plt.plot(x, y)

    r = np.where((np.isclose(x, 0.0)) & (np.isclose(y, 0.0)))[0]
    
    if len(r): return r[0]
    else : return steps
    
sims = np.array([make_walk(steps) for i in range(walks)])
```

```python
returned = sims[np.where(sims < 100)]
len(returned) / walks
```

```python
w, s = np.unique(sims, return_counts=True)
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.bar(w+1, s/s.sum())
plt.subplot(1,2,2)
plt.bar(w[:10]+1, (s/s.sum())[:10], width=0.2)
```

```python
w, s = np.unique(returned, return_counts=True)
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.bar(w+1, s/s.sum())
plt.subplot(1,2,2)
plt.bar(w[:10]+1, (s/s.sum())[:10], width=0.1)
```
